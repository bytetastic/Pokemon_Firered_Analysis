### Jetzt Typ-Bereinigung
### Welche Pokemon haben Typ1 == Typ2 ? ⬇️

```python
gleiche_typen = df_pokemon[df_pokemon["typ1"] == df_pokemon["typ2"]]
print(f"Pokemon mit nur einem Typ: {len(gleiche_typen)}")
gleiche_typen
```

### Ausgabe mit len zeigt 87 x Typ1 == Typ2
### Ausgabe mit Variable "gleiche_typen" zeigt jedoch 86 unique Values an
### Test welches evtl. doppelt ist ⬇️

```python
print(gleiche_typen[gleiche_typen.duplicated(subset="name", keep=False)][["name", "typ1", "typ2"]])
```

### Nidoran männlich und weiblich haben normalerweise Geschlechtszeichen neben dem Namen https://www.bisafans.de/pokedex/029.php#attacken https://www.bisafans.de/pokedex/032.php
### Passe Nidoran mit W und M an ⬇️

```python
df_pokemon.loc[df_pokemon.index == 29, "name"] = "NIDORAN_W"
df_pokemon.loc[df_pokemon.index == 32, "name"] = "NIDORAN_M"
```

```python
df_pokemon.loc[[29, 32], ["name", "typ1", "typ2"]]
```

### Nidoran männlich und weiblich Namen angepasst
### Kurzer Check ⬇️

```python
gleiche_typen_neu = df_pokemon[df_pokemon["typ1"] == df_pokemon["typ2"]]

duplikate = gleiche_typen_neu[gleiche_typen_neu.duplicated(subset="name", keep=False)]

print(gleiche_typen_neu[gleiche_typen_neu.duplicated(subset="name", keep=False)][["name", "typ1", "typ2"]])
```

### Als nächstes Typ1 == Typ2 anpassen
### Wenn Typ1 == Typ2 dann Typ2 auf None setzen ⬇️

```python
df_pokemon.loc[df_pokemon["typ1"] == df_pokemon["typ2"], "typ2"] = None
```

```python
print(f"Pokemon mit zwei Typen: {df_pokemon['typ2'].notna().sum()}")
print(f"Pokemon mit einem Typ:  {df_pokemon['typ2'].isna().sum()}")
```

### Ausgabe: Pokemon mit einem Typ 87 passt
### Jetzt unbekannte Typen bei den Moves bereinigen ⬇️

```python
unbekannte_moves = df_moves[df_moves["typ"].str.contains("Unbekannt")]
print(f"Moves mit unbekanntem Typ: {len(unbekannte_moves)}")
print(unbekannte_moves[["move_name", "typ", "staerke"]])
```

### Keine unbekannten Moves mehr?
